```python
import numpy as np
import pandas as pd
PGA_TechData = pd.read_csv('../data/PGATOUR_data2.csv')
PGA_GenData = pd.read_csv('../data/pgaTourData.csv')
```

```python
PGA_GenData
```

```python
PGA_TechData
```

```python
# merge two datasets together, need information include all technical data and numbers that could indicate players' overall performance
PGA_DATA = pd.merge(PGA_TechData, PGA_GenData[['Player Name', 'Year', 'Average Score', 'Wins', 'Top 10', 'Money']], left_on='Player', right_on='Player Name')
# cleanning up duplicated data
del PGA_DATA['Player Name']
del PGA_DATA['POINTS']
del PGA_DATA['NUMBER_OF_WINS']
# cleanning up unneeded data
del PGA_DATA['NUMBER_OF_TOP_Tens']
del PGA_DATA['POINTS_BEHIND_LEAD']
del PGA_DATA['EVENTS_PLAYED']
# show the dataset
PGA_DATA
```

```python
import pandas.plotting._matplotlib as module
# Only looking at data from year 2018 since it is the latest and thus the most relevent data.
temp1 = PGA_DATA[PGA_DATA.Year == 2018]
temp1.plot(x = 'SG_PUTTING_PER_ROUND', y = 'Average Score', kind = 'scatter')
```
